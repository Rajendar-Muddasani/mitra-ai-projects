## DQN Key Innovations

1. **Experience Replay**: store (s,a,r,s') in buffer, sample random mini-batches
2. **Target Network**: separate slow-updating network stabilises training
3. **Epsilon-greedy**: balance exploration vs exploitation

````python
import numpy as np, matplotlib.pyplot as plt
np.random.seed(42)
rewards = np.cumsum(np.random.randn(200).clip(-1,1))
plt.figure(figsize=(8,3))
plt.plot(rewards, color='#6b21a8', lw=1.5)
plt.xlabel('Episode'); plt.ylabel('Cumulative Reward')
plt.title('Simulated RL Training Progress')
plt.grid(alpha=0.3); plt.show()
print("Run stable-baselines3 on Kaggle for real RL training!")
````
