## The 5 Principles

1. **Choose the right chart** -- not all data suits a bar chart
2. **Remove chart junk** -- eliminate anything that doesn't help the reader
3. **Pre-attentive attributes** -- use colour and position to guide the eye
4. **Direct labelling** -- labels > legends when possible
5. **Tell one story per chart** -- don't try to show everything at once

```python
import matplotlib.pyplot as plt, numpy as np

fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(13, 5))
months = ['Jan','Feb','Mar','Apr','May','Jun']
revenue = [82, 91, 88, 105, 120, 135]
highlight = ['#6b21a8' if r == max(revenue) else '#d1b3f5' for r in revenue]

# BAD VERSION
ax1.bar(months, revenue, color='steelblue', width=0.8)
ax1.set_title('Monthly Revenue', fontsize=12)
ax1.set_ylabel('Revenue'); ax1.grid(True); ax1.set_xlabel('Month')
ax1.text(0.5, -0.18, 'BAD: cluttered, no emphasis, no insight',
         transform=ax1.transAxes, ha='center', color='#dc2626', fontsize=9)

# GOOD VERSION
ax2.bar(months, revenue, color=highlight, width=0.65)
ax2.set_title('Revenue peaked in June -- H1 target exceeded', fontsize=11, fontweight='bold', pad=12)
for i, (m, r) in enumerate(zip(months, revenue)):
    ax2.text(i, r+1.5, f'Rs{r}K', ha='center', va='bottom', fontsize=9,
             fontweight='bold' if r==max(revenue) else 'normal',
             color='#6b21a8' if r==max(revenue) else '#555')
ax2.annotate('+64% vs Jan', xy=(5,135), xytext=(3.5,128),
             arrowprops=dict(arrowstyle='->', color='#059669'),
             color='#059669', fontsize=9, fontweight='bold')
ax2.set_ylabel('Revenue (Rs Thousands)', fontsize=10)
ax2.spines['top'].set_visible(False); ax2.spines['right'].set_visible(False)
ax2.grid(axis='y', alpha=0.2)
ax2.text(0.5, -0.18, 'GOOD: insight in title, emphasis, minimal clutter',
         transform=ax2.transAxes, ha='center', color='#059669', fontsize=9)

plt.suptitle('Bad vs Good Visualisation', fontsize=13, fontweight='bold')
plt.tight_layout(pad=2); plt.show()
```

```python
chart_guide = {
    'Bar chart':       'Comparing categories (few items)',
    'Line chart':      'Trends over time (continuous)',
    'Scatter plot':    'Relationship between 2 numeric variables',
    'Histogram':       'Distribution of one numeric variable',
    'Heatmap':         'Matrix of values (correlations, pivot tables)',
    'Box plot':        'Distribution comparison across groups',
    'Pie chart':       'Part-of-whole (max 5 slices -- rarely best choice)',
    'Waterfall':       'Cumulative additions/subtractions',
    'Geographic map':  'Data with a location dimension',
}
print("CHART SELECTION GUIDE:")
print("="*55)
for chart, use in chart_guide.items():
    print(f"  {chart:<22} -- {use}")

print("\nGOLDEN RULES:")
for rule in [
    "Never use 3D charts -- they distort proportions",
    "Bar charts must start at zero",
    "Use colour to highlight ONE thing, not decorate everything",
    "Put the insight in the chart title, not just the variable name",
    "Pie charts only when parts sum to 100%, max 5 slices",
]:
    print(f"  + {rule}")
```

```python
# The IBCS (International Business Communication Standards) colour rules
categories = ['Revenue','Cost','Margin','Target','Forecast']
actuals  = [150, 80, 70, 140, 160]
colors_wrong = ['blue','red','green','orange','purple']
colors_right = ['#6b21a8','#dc2626','#6b21a8','#aaaaaa','#aaaaaa']
fig, (ax1,ax2) = plt.subplots(1,2,figsize=(11,4))
ax1.bar(categories, actuals, color=colors_wrong)
ax1.set_title('Rainbow = confusing (what does each colour mean?)', fontsize=10)
ax2.bar(categories, actuals, color=colors_right)
ax2.set_title('Semantic colour: purple=actual, grey=reference', fontsize=10)
for ax in [ax1,ax2]:
    ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
plt.tight_layout(); plt.show()
```
