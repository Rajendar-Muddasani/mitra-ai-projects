## Tokenisation
LLMs read **tokens** (subword units), not characters or words.

```python
# Token count estimation
text = 'Machine learning is transforming how we build software applications today.'
words = text.split()
print(f'Words: {len(words)}')
print(f'Approx GPT-4 tokens: ~{len(text)//4}')
print()
examples = [('the',1),('machine',1),('tokenisation',3),('applications',4),('antidisestablishmentarianism',9)]
for word, count in examples:
    bar = '#' * count
    print(f'  {word:<30} {bar} ({count} tokens)')
```

## LLM Cost Estimation

```python
# Cost per 1000 API calls (500 input + 200 output tokens)
pricing = {
    'GPT-4o':           (2.50, 10.00),
    'GPT-4o-mini':      (0.15, 0.60),
    'Claude 3.5 Sonnet':(3.00, 15.00),
    'Claude 3 Haiku':   (0.25, 1.25),
    'Gemini 1.5 Flash': (0.075,0.30),
}
print(f'{'Model':<25}| $/1K calls')
print('-'*40)
for model, (inp, out) in pricing.items():
    cost = inp*0.5 + out*0.2
    print(f'{model:<25}| ${cost:.4f}')
```

## Temperature and Sampling

```python
import numpy as np, matplotlib.pyplot as plt
logits = np.array([3.0, 1.5, 0.8, 0.3, -0.5, -1.2, -2.0, -3.0])
def softmax(x, t=1.0):
    e = np.exp((x - x.max()) / t)
    return e / e.sum()
temps = [0.1, 0.7, 1.0, 1.5, 2.0]
fig, axes = plt.subplots(1,5,figsize=(14,3))
for ax,t in zip(axes,temps):
    ax.bar(range(len(logits)),softmax(logits,t),color='#6b21a8')
    ax.set_title(f'T={t}',fontsize=10); ax.set_ylim(0,1)
plt.suptitle('Effect of Temperature on Token Probabilities')
plt.tight_layout(); plt.show()
print('Low temp: deterministic. High temp: creative/varied.')
```

## Discriminative vs Generative

| | Discriminative | Generative |

| Goal | Classify/predict | Create new content |

| Examples | SVM, Random Forest | GPT, DALL-E |
| Training | Supervised | Self-supervised |
