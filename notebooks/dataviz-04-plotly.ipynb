## Why Plotly?

- Interactive: hover, zoom, pan, click in the browser
- Export to HTML -- shareable without a server
- Plotly Express: high-level (like seaborn)
- Graph Objects: full control
- Dash: full web dashboards

```python
import plotly.express as px
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import pandas as pd, numpy as np
np.random.seed(42)
n = 300
df = pd.DataFrame({
    'country':  np.random.choice(['India','USA','UK','Germany','Japan','Brazil'], n),
    'revenue':  np.random.exponential(50, n) * 1000,
    'users':    np.random.randint(100, 10000, n),
    'growth':   np.random.normal(0.15, 0.12, n),
    'year':     np.random.choice([2022,2023,2024,2025], n),
    'category': np.random.choice(['SaaS','E-commerce','Fintech','EdTech'], n),
})
print(df.shape)
```

```python
# Interactive scatter -- hover for details
fig = px.scatter(df, x='users', y='revenue',
    color='category', size='revenue',
    hover_data=['country','growth','year'],
    title='Revenue vs Users (hover for details, click legend to filter)',
    labels={'users':'Monthly Active Users', 'revenue':'Revenue ($)'},
    template='plotly_white',
    color_discrete_sequence=px.colors.qualitative.Set2)
fig.update_traces(marker=dict(opacity=0.75, line=dict(width=1, color='white')))
fig.update_layout(height=450)
fig.show()
```

```python
# Animated bar chart
yearly = df.groupby(['year','country','category']).agg(
    revenue=('revenue','sum'), users=('users','sum')).reset_index()

fig = px.bar(yearly, x='country', y='revenue', color='category',
    animation_frame='year',
    title='Revenue by Country and Category (press Play)',
    template='plotly_white',
    color_discrete_sequence=px.colors.qualitative.Set1)
fig.update_layout(height=450)
fig.show()
```

```python
# Multi-panel dashboard
fig = make_subplots(rows=1, cols=3,
    subplot_titles=['Revenue Distribution','Users by Country','Growth Rate'])

fig.add_trace(go.Histogram(x=df['revenue'], nbinsx=30,
    marker_color='#6b21a8', opacity=0.8, name='Revenue'), row=1, col=1)

for country in df['country'].unique():
    fig.add_trace(go.Box(y=df[df['country']==country]['users'],
        name=country, showlegend=False), row=1, col=2)

fig.add_trace(go.Violin(y=df['growth'], box_visible=True,
    fillcolor='rgba(107,33,168,0.3)', line_color='#6b21a8',
    name='Growth', showlegend=False), row=1, col=3)

fig.update_layout(height=420, title_text='Business Metrics Dashboard',
    template='plotly_white', showlegend=False)
fig.show()
```
