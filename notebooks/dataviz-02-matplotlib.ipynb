## Matplotlib Architecture

```
Figure --> Axes --> Artists (lines, patches, text)
```

- `Figure` = the whole image canvas
- `Axes` = one plot area
- Every visual = an `Artist`

````python
import matplotlib.pyplot as plt, numpy as np

fig, ax = plt.subplots(figsize=(9, 5))
x = np.linspace(0, 4*np.pi, 300)
ax.plot(x, np.sin(x), color='#6b21a8', lw=2, label='sin(x)')
ax.plot(x, np.cos(x), color='#059669', lw=2, linestyle='--', label='cos(x)')
ax.fill_between(x, np.sin(x), alpha=0.12, color='#6b21a8')
ax.set_title('Sine and Cosine Waves', fontsize=14, fontweight='bold', pad=15)
ax.set_xlabel('x (radians)', fontsize=11)
ax.set_ylabel('Amplitude', fontsize=11)
ax.legend(framealpha=0.9, fontsize=10)
ax.grid(True, alpha=0.3, linestyle=':')
ax.set_xlim(0, 4*np.pi)
ax.set_xticks([0, np.pi, 2*np.pi, 3*np.pi, 4*np.pi])
ax.set_xticklabels(['0', 'pi', '2pi', '3pi', '4pi'])
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
plt.tight_layout(); plt.show()
````

````python
import matplotlib.gridspec as gridspec

fig = plt.figure(figsize=(12, 6))
gs = gridspec.GridSpec(2, 3, figure=fig, hspace=0.4, wspace=0.35)
np.random.seed(42)
data = np.random.randn(500)
x_s = np.random.randn(100)
y_s = x_s * 0.7 + np.random.randn(100) * 0.5

ax1 = fig.add_subplot(gs[:, 0:2])
ax1.hist(data, bins=25, color='#6b21a8', alpha=0.8, edgecolor='white')
ax1.set_title('Distribution', fontweight='bold')

ax2 = fig.add_subplot(gs[0, 2])
ax2.scatter(x_s, y_s, s=15, color='#059669', alpha=0.7)
ax2.set_title('Scatter', fontsize=9)

ax3 = fig.add_subplot(gs[1, 2])
ax3.boxplot(data, patch_artist=True, boxprops=dict(facecolor='#f3e8ff', color='#6b21a8'))
ax3.set_title('Boxplot', fontsize=9)

plt.suptitle('Multi-Plot Dashboard', fontsize=14, fontweight='bold')
plt.show()
````

````python
# Annotated line chart
months = ['Jan','Feb','Mar','Apr','May','Jun','Jul','Aug','Sep','Oct','Nov','Dec']
revenue = [82, 91, 88, 105, 120, 135, 128, 142, 155, 148, 162, 178]

fig, ax = plt.subplots(figsize=(11, 5))
ax.plot(months, revenue, 'o-', color='#6b21a8', lw=2.5, markersize=7,
        markerfacecolor='white', markeredgewidth=2)
ax.fill_between(range(12), revenue, alpha=0.1, color='#6b21a8')

max_idx = revenue.index(max(revenue))
ax.annotate(f'Peak: Rs{max(revenue)}K',
            xy=(max_idx, max(revenue)), xytext=(max_idx-2, max(revenue)+10),
            arrowprops=dict(arrowstyle='->', color='#dc2626', lw=1.5),
            fontsize=10, color='#dc2626', fontweight='bold')

ax.set_title('Monthly Revenue 2026 -- Best year yet!', fontsize=13, fontweight='bold')
ax.set_ylabel('Revenue (Rs Thousands)')
ax.grid(axis='y', alpha=0.3)
ax.spines['top'].set_visible(False); ax.spines['right'].set_visible(False)
plt.tight_layout(); plt.show()
````
