```python
# Variables and types
name  = "Mitra AI"
score = 95.5
items = [10, 20, 30, 40, 50]
info  = {"language": "Python", "year": 2024}

print(type(name), type(score), type(items), type(info))
print(f"Name: {name}, Score: {score}")
```

```python
# List comprehension
squares   = [x**2 for x in range(1, 11)]
even_sq   = [x**2 for x in range(1, 11) if x % 2 == 0]
print("Squares:", squares)
print("Even squares:", even_sq)
```

```python
# Functions
def greet(name, greeting="Hello"):
    return f"{greeting}, {name}!"

print(greet("Arjun"))
print(greet("Priya", "Welcome"))

# Lambda
double = lambda x: x * 2
print(list(map(double, [1, 2, 3, 4])))
```

```python
# Exception handling
def safe_divide(a, b):
    try:
        return a / b
    except ZeroDivisionError:
        return "Cannot divide by zero"

print(safe_divide(10, 2))
print(safe_divide(10, 0))
```
