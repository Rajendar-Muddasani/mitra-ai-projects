# Perl

Perl regex, file processing, and scripting — the language behind legacy bioinformatics, sysadmin, and data pipelines.

Perl doesn't run in the browser, so we'll use Python to demonstrate the same patterns.

```python
import re

# Perl regex ≈ Python re module
text = "Name: Arjun Kumar, Phone: 98765-43210, Email: arjun@example.com"

# Pattern matching (Perl: $text =~ /pattern/)
match = re.search(r'Phone: ([\d-]+)', text)
if match:
    print("Phone:", match.group(1))

# Global find all (Perl: while ($text =~ /pattern/g))
emails = re.findall(r'[\w.]+@[\w.]+', text)
print("Emails found:", emails)
```

```python
# String splitting (Perl: split /,/, $str)
csv_line = "Alice,CSE,9.1,Placed"
fields = csv_line.split(',')
print("Fields:", fields)
print("Name:", fields[0], "CGPA:", fields[2])
```

```python
# File processing pattern (Perl: while(<FILE>))
# We'll simulate reading lines
lines = [
    "student: Alice, cgpa: 9.1",
    "student: Bob, cgpa: 7.5",
    "student: Charlie, cgpa: 8.8",
]

# Perl: while ($line = <FILE>) { chomp $line; ... }
# Python equivalent:
for line in lines:
    line = line.strip()  # chomp equivalent
    match = re.search(r'student: (\w+), cgpa: ([\d.]+)', line)
    if match:
        name, cgpa = match.group(1), float(match.group(2))
        if cgpa >= 8.5:
            print(f"High achiever: {name} ({cgpa})")
```

**Perl → Python cheat sheet:**

| Perl | Python |
|---|---|
| `$scalar` | `variable` |
| `@array` | `list` |
| `%hash` | `dict` |
| `chomp $str` | `str.strip()` |
| `split /,/, $str` | `str.split(',')` |
| `$str =~ /pattern/` | `re.search(pattern, str)` |
