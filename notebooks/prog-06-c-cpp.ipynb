# C / C++ Concepts

Key C/C++ concepts illustrated with Python equivalents. C itself doesn't run in the browser, but understanding the logic does.

> Use this notebook to understand pointers, memory, and structs conceptually — then apply in your C/C++ IDE.

```python
# Pointer concept in Python (using id() and references)
a = 42
b = a  # b gets a copy of the value
print(f"a = {a}, id(a) = {id(a)}")
print(f"b = {b}, id(b) = {id(b)}")

# In Python, small ints are cached, so id is same
# In C: int a=42; int *ptr=&a; — ptr holds the address of a
```

```python
# Dynamic memory (C malloc equivalent)
# In C: int *arr = malloc(5 * sizeof(int));
# In Python: just use a list (memory managed automatically)
arr = [0] * 5
print("Allocated array:", arr)

# Filling it
for i in range(5):
    arr[i] = i * 10
print("Filled array:", arr)

# In C you must free: free(arr)
# In Python: garbage collector handles it
```

```python
# Struct equivalent using Python dataclass
from dataclasses import dataclass

@dataclass
class Student:
    name: str
    roll: int
    cgpa: float

# In C:
# struct Student { char name[50]; int roll; float cgpa; };

s1 = Student("Arjun", 101, 9.1)
s2 = Student("Priya", 102, 8.5)
print(s1)
print(f"Name: {s2.name}, CGPA: {s2.cgpa}")
```

```python
# Recursion — same in C and Python
def factorial(n):
    if n <= 1:
        return 1
    return n * factorial(n - 1)

for i in range(1, 8):
    print(f"{i}! = {factorial(i)}")
```

**C/C++ concept summary:**

| Concept | C syntax | Python equivalent |
|---|---|---|
| Pointer | `int *ptr = &x` | `id(x)` (memory address) |
| Malloc | `malloc(n * sizeof(int))` | `[0] * n` |
| Struct | `struct Student {...}` | `@dataclass class Student` |
| Free | `free(ptr)` | automatic (GC) |
