```python
import pandas as pd
import numpy as np

# Create a sample DataFrame
data = {
    'student': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
    'branch':  ['CSE', 'ECE', 'CSE', 'IT', 'CSE'],
    'cgpa':    [8.9, 7.5, 9.1, 8.2, 7.8],
    'placed':  [True, False, True, True, False]
}
df = pd.DataFrame(data)
print(df)
print("\nShape:", df.shape)
print("\nDescribe:\n", df.describe())
```

```python
# Filtering and groupby
cse = df[df['branch'] == 'CSE']
print("CSE students:\n", cse)

branch_avg = df.groupby('branch')['cgpa'].mean()
print("\nBranch avg CGPA:\n", branch_avg)
```

```python
import matplotlib.pyplot as plt

plt.bar(branch_avg.index, branch_avg.values, color=['purple','teal','orange'])
plt.xlabel('Branch'); plt.ylabel('Average CGPA')
plt.title('Average CGPA by Branch'); plt.show()
```

```python
# NumPy operations
arr = np.array([8.9, 7.5, 9.1, 8.2, 7.8])
print("Mean:", arr.mean().round(2))
print("Std: ", arr.std().round(2))
print("Normalised:", ((arr - arr.mean()) / arr.std()).round(2))
```
