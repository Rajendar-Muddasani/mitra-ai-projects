```python
import pandas as pd
import numpy as np

# Sample data (like an Excel sheet)
df = pd.DataFrame({
    'Name':   ['Alice','Bob','Charlie','Diana','Eve','Frank'],
    'Dept':   ['HR','IT','IT','Finance','HR','IT'],
    'Salary': [45000, 75000, 80000, 60000, 50000, 72000],
    'Target': [40000, 70000, 75000, 55000, 45000, 68000],
    'Joined': [2019, 2020, 2018, 2021, 2022, 2020],
})
print(df)
```

```python
# VLOOKUP equivalent: merge/map
lookup_table = pd.DataFrame({'Dept': ['HR','IT','Finance'], 'Dept_Code': ['H01','I01','F01']})
df_merged = df.merge(lookup_table, on='Dept', how='left')
print("With Dept Code (VLOOKUP equivalent):\n", df_merged[['Name','Dept','Dept_Code']])
```

```python
# SUMIFS, COUNTIFS, AVERAGEIFS equivalents
it_mask = df['Dept'] == 'IT'
print("SUMIFS   — IT Salary Total :", df.loc[it_mask, 'Salary'].sum())
print("COUNTIFS — IT employees    :", it_mask.sum())
print("AVERAGEIF— IT avg salary   :", df.loc[it_mask, 'Salary'].mean().round(0))

# IF equivalent
df['Bonus'] = np.where(df['Salary'] > df['Target'], 'Bonus Eligible', 'No Bonus')
print("\nIF formula equivalent:\n", df[['Name','Salary','Target','Bonus']])
```

```python
# Pivot table equivalent
pivot = df.pivot_table(values='Salary', index='Dept', aggfunc=['mean','count','sum'])
print("Pivot Table:\n", pivot.round(0))
```
