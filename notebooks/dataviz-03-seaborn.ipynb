## Why Seaborn?

- Built-in confidence intervals and statistical summaries
- Prettier defaults than raw matplotlib
- Easy categorical and distribution plots
- FacetGrid for multi-panel by group

```python
import seaborn as sns, matplotlib.pyplot as plt, numpy as np, pandas as pd
sns.set_theme(style='whitegrid', palette='husl')
np.random.seed(42)
n = 400
df = pd.DataFrame({
    'score':    np.concatenate([np.random.normal(72,12,n//2), np.random.normal(80,10,n//2)]),
    'group':    ['Control']*(n//2) + ['Treatment']*(n//2),
    'gender':   np.random.choice(['M','F'], n),
    'subject':  np.random.choice(['Math','Science','English','History'], n),
    'hours':    np.random.gamma(2, 5, n).clip(0, 40),
})
print(df.shape); print(df.head(3))
```

```python
fig, axes = plt.subplots(2, 3, figsize=(14, 9))

sns.histplot(data=df, x='score', hue='group', kde=True, ax=axes[0,0], alpha=0.6, palette=['#6b21a8','#059669'])
axes[0,0].set_title('Score Distribution by Group')

sns.boxplot(data=df, x='subject', y='score', hue='group', ax=axes[0,1], palette='Set2')
axes[0,1].set_title('Score by Subject and Group')
axes[0,1].tick_params(axis='x', rotation=20)

sns.violinplot(data=df, x='gender', y='score', inner='quartile', ax=axes[0,2], palette='pastel')
axes[0,2].set_title('Score Distribution by Gender')

sns.regplot(data=df, x='hours', y='score', ax=axes[1,0],
            scatter_kws={'alpha':0.3,'s':15}, line_kws={'color':'#dc2626','lw':2})
axes[1,0].set_title('Study Hours vs Score')

pivot = df.pivot_table(values='score', index='gender', columns='subject', aggfunc='mean')
sns.heatmap(pivot, annot=True, fmt='.0f', cmap='RdYlGn', ax=axes[1,1], vmin=60, vmax=90)
axes[1,1].set_title('Mean Score Heatmap')

sns.countplot(data=df, x='subject', hue='group', ax=axes[1,2], palette='Set1')
axes[1,2].set_title('Sample Count by Subject')

plt.suptitle('Student Performance Dashboard', fontsize=14, fontweight='bold', y=1.01)
plt.tight_layout(); plt.show()
```

```python
# FacetGrid -- multi-panel by category
g = sns.FacetGrid(df, col='subject', hue='group', height=3.5, aspect=0.9, col_wrap=4)
g.map_dataframe(sns.kdeplot, x='score', fill=True, alpha=0.4)
g.add_legend()
g.set_titles('{col_name}')
g.fig.suptitle('Score Distribution per Subject', y=1.03, fontsize=13, fontweight='bold')
plt.show()
```
