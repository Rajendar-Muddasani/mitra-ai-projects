# SQL

Query, join, aggregate, and analyse data — using Python's built-in `sqlite3` module.

```python
import sqlite3
import pandas as pd

# Create in-memory database
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()

# Create tables
cursor.executescript('''
CREATE TABLE students (id INT, name TEXT, branch TEXT, cgpa REAL, placement_year INT);
CREATE TABLE companies (id INT, name TEXT, package_lpa REAL);
CREATE TABLE placements (student_id INT, company_id INT);

INSERT INTO students VALUES
  (1,'Arjun','CSE',9.1,2024),(2,'Priya','IT',8.5,2024),
  (3,'Rahul','ECE',7.8,2023),(4,'Sneha','CSE',9.4,2024),
  (5,'Kiran','IT',7.2,2023);
INSERT INTO companies VALUES
  (1,'TCS',7.5),(2,'Infosys',8.0),(3,'Wipro',6.5),(4,'Amazon',18.0);
INSERT INTO placements VALUES (1,4),(2,2),(3,1),(4,4),(5,3);
''')
conn.commit()
print("Database created with students, companies, placements")
```

```python
# SELECT with WHERE
df = pd.read_sql("SELECT * FROM students WHERE cgpa > 8.0 ORDER BY cgpa DESC", conn)
print("High CGPA students:\n", df)
```

```python
# GROUP BY with HAVING
sql = "SELECT branch, COUNT(*) as count, ROUND(AVG(cgpa),2) as avg_cgpa FROM students GROUP BY branch HAVING count > 1"
df = pd.read_sql(sql, conn)
print("Branch summary:\n", df)
```

```python
# JOIN — students with their company offers
sql = """
    SELECT s.name, s.branch, s.cgpa, c.name as company, c.package_lpa
    FROM students s
    JOIN placements p ON s.id = p.student_id
    JOIN companies c  ON c.id = p.company_id
    ORDER BY c.package_lpa DESC
"""
df = pd.read_sql(sql, conn)
print("Placement details:\n", df)
```
