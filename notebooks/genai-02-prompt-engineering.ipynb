## Core Patterns

| Pattern | When to use |
|---|---|
| Zero-shot | Simple tasks, capable models |
| Few-shot | Custom formats, domain-specific tasks |
| Chain-of-Thought | Math, logic, multi-step reasoning |
| ReAct | Tool use, agents, planning |
| Role prompting | Expert-level domain framing |

````python
patterns = {
    'Zero-shot':  'Classify: "{review}" -> Sentiment:',
    'Few-shot':   '"Great!" -> Positive\n"Awful" -> Negative\n"{review}" ->',
    'CoT':        'Q: 60km/h for 2h then 30min stop? A: Let me think step by step...',
    'ReAct':      'Thought: need capital of France\nAction: search()\nObservation: Paris',
    'Role':       'You are a senior ML engineer. Review this code for performance issues:',
}
for name, template in patterns.items():
    print(f'{name}:')
    print(f'  {template[:80]}...')
    print()
````

## System Prompt Structure

```
You are {persona}.
You help {audience} with {task}.
Always {constraint1}. Never {constraint2}.
Respond in {format}. Keep answers under {length}.
```

````python
ml_tutor = [
    'You are an expert ML tutor with deep knowledge of sklearn, PyTorch, and statistics.',
    'You help engineering students understand ML concepts and debug their code.',
    'Always explain with a real-world analogy first, then the technical definition.',
    'Never give full homework solutions -- guide the student to find the answer.',
    'Keep explanations under 200 words. Use code blocks for all code.',
]
print('ML Tutor System Prompt:')
for line in ml_tutor:
    print(f'  {line}')
````

## Anti-Patterns

| Anti-pattern | Bad | Good |
|---|---|---|
| Vague | 'Write about ML' | 'Write 3 paragraphs on supervised ML for CS students' |
| No format | 'List algorithms' | 'Table: Name | Use case | Key hyperparameter' |
| Ambiguous | 'Make it short' | 'Answer in exactly 2 sentences' |
| Leading | 'Isn't NN best?' | 'What approach do you recommend and why?' |

````python
# RAG-style context injection (simulate what happens before the LLM call)
docs = [
    'Linear regression fits y = wx + b by minimising MSE.',
    'Ridge adds L2 penalty alpha*sum(w^2) to prevent large weights.',
    'Lasso adds L1 penalty which zeros out some weights.',
]
question = 'What is the difference between Ridge and Lasso?'
context = '\n'.join(f'[Doc {i+1}]: {d}' for i,d in enumerate(docs))
print('System prompt with injected context:')
print(f'  Context: {context[:100]}...')
print(f'  Question: {question}')
````
