````python
import numpy as np

class FakeRepo:
    def __init__(self, name):
        self.name = name
        self.commits = []
        self.branch = 'main'
    def add(self, files):
        print(f"Staged: {files if isinstance(files,list) else [files]}")
    def commit(self, msg):
        self.commits.append(msg)
        print(f"Commit [{len(self.commits)}]: {msg}")
    def checkout(self, branch, b=False):
        self.branch = branch
        print(f"On branch: {branch}")

repo = FakeRepo('ml-project')
repo.add(['src/train.py', 'requirements.txt'])
repo.commit("feat: initial training pipeline")
repo.checkout('experiment/xgboost', b=True)
repo.add('configs/params.yaml')
repo.commit("exp: try XGBoost hyperparameters")
repo.checkout('main')
print(f"\nCommit history: {repo.commits}")
````
