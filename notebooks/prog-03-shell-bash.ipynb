# Shell / Bash — Python Simulation

Common shell operations demonstrated with Python's `os` and `subprocess` modules.

> Shell commands themselves don't run in the browser — but the Python equivalents do, and knowing both is practical.

```python
import os

# os.listdir ≈ ls
files = os.listdir('.')
print("Files in current directory:")
for f in files[:10]:
    print(' ', f)
```

```python
# os.path ≈ pwd, basename, dirname
cwd = os.getcwd()
print("Current directory:", cwd)

path = "/home/user/projects/model.py"
print("basename:", os.path.basename(path))   # filename
print("dirname: ", os.path.dirname(path))    # directory
print("exists:  ", os.path.exists(path))     # file check
```

```python
# Writing and reading files ≈ echo > file, cat file
with open('/tmp/test.txt', 'w') as f:
    f.write("line 1\nline 2\nline 3\n")

with open('/tmp/test.txt') as f:
    content = f.read()
print(content)
```

```python
# grep equivalent with Python
lines = ["import pandas", "import numpy", "from sklearn import tree", "# this is a comment"]
keyword = "import"
matches = [l for l in lines if keyword in l]
print(f"Lines containing '{keyword}':")
for m in matches:
    print(' ', m)
```

**Key command mapping:**

| Shell | Python |
|---|---|
| `ls` | `os.listdir()` |
| `pwd` | `os.getcwd()` |
| `cat` | `open().read()` |
| `grep` | list comprehension with `in` |
| `wc -l` | `len(lines)` |
