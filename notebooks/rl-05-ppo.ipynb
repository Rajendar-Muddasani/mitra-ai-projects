## Why PPO?

- Stable training (clips large policy updates)
- Works well out of the box
- Used in: InstructGPT/ChatGPT RLHF, game playing, robotics

```python
from stable_baselines3 import PPO
model = PPO('MlpPolicy', 'CartPole-v1', verbose=1)
model.learn(total_timesteps=100_000)
```

````python
import numpy as np, matplotlib.pyplot as plt
np.random.seed(42)
rewards = np.cumsum(np.random.randn(200).clip(-1,1))
plt.figure(figsize=(8,3))
plt.plot(rewards, color='#6b21a8', lw=1.5)
plt.xlabel('Episode'); plt.ylabel('Cumulative Reward')
plt.title('Simulated RL Training Progress')
plt.grid(alpha=0.3); plt.show()
print("Run stable-baselines3 on Kaggle for real RL training!")
````
